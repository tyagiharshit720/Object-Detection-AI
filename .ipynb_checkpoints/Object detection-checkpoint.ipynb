```python
import cv2
```

```python
import pygame
```

```python
import matplotlib.pyplot as plt
```

```python
#ModelConfiguration
config_file = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model = 'frozen_inference_graph.pb'
```

```python
#ModelLoading
model = cv2.dnn_DetectionModel(frozen_model,config_file)
```

```python
classLabels= []
file_name= 'Labels.txt'
with open(file_name,'rt') as fpt:
    classLabels=fpt.read().rstrip('\n').split('\n')
```

```python
print(classLabels)
```

```python
print(len(classLabels))
```

```python
model.setInputSize(320,320) 
model.setInputScale(1.0/127.5) 
model.setInputMean ((127.5, 127.5,127.5)) 
model.setInputSwapRB (True)
```

```python
img= cv2.imread('boy.jpg')
plt.imshow(img)
```

```python
ClassIndex, confidece, bbox = model.detect(img, confThreshold=0.5)
```

```python
print(ClassIndex)
```

```python
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN

for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(), bbox):
    cv2.rectangle(img, boxes, (255, 0, 0), 2) 
    cv2.putText(img, classLabels[ClassInd-1], (boxes[0]+10, boxes[1]+40), font, fontScale=font_scale, color=(0, 255, 0), thickness=3)

```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
#Video 

cap = cv2.VideoCapture("video_Sample.MP4")

# Check if the video is opened correctly
if not cap.isOpened():
    cap = cv2.VideoCapture(0)

if not cap.isOpened():
    raise IOError("Can't open the video")

font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN

# Initialize the Pygame mixer
pygame.mixer.init()


# Load the audio file
audio_file = r"C:\Users\tyagi\OneDrive\Desktop\Object Detection AI\horn_sound.MP3"

pygame.mixer.music.load(audio_file)

# Classes you want to detect
desired_classes = ["person", "bicycle", "car", "motorbike", "Mobile Phone", "keyboard", "laptop", "mouse"]

while True:
    ret, frame = cap.read()

    classIndex, confidence, bbox = model.detect(frame, confThreshold=0.55)

    # Stop the beep when no desired class is detected
    pygame.mixer.music.stop()

    if (len(classIndex) != 0):
        for ClassInd, conf, boxes in zip(classIndex.flatten(), confidence.flatten(), bbox):
            if ClassInd <= len(classLabels):
                class_name = classLabels[ClassInd - 1]

                if class_name in desired_classes:
                    cv2.rectangle(frame, boxes, (255, 0, 0), 2)
                    cv2.putText(frame, class_name, (boxes[0] + 10, boxes[1] + 40),
                                font, fontScale=font_scale, color=(0, 255, 0), thickness=3)

                    # Play the continuous beep when a desired class is detected
                    pygame.mixer.music.play(-1)  # -1 indicates continuous play

    cv2.imshow('Object Detection AI', frame)

    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()






```
